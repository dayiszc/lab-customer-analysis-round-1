```python
import pandas as pd

url1 = "https://github.com/ironhack-labs/lab-customer-analysis-round-1/raw/master/files_for_lab/csv_files/file1.csv"
url2 = "https://github.com/ironhack-labs/lab-customer-analysis-round-1/raw/master/files_for_lab/csv_files/file2.csv"
url3 = "https://github.com/ironhack-labs/lab-customer-analysis-round-1/raw/master/files_for_lab/csv_files/file3.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3) 
```

```python
print("Shape of df1:", df1.shape)
print("Shape of df2:", df2.shape)
print("Shape of df3:", df3.shape) #DataFrame's shape
```

```python
df1.columns = df1.columns.str.lower().str.replace(' ', '_')
df2.columns = df2.columns.str.lower().str.replace(' ', '_')
df3.columns = df3.columns.str.lower().str.replace(' ', '_') # Standardize header names
```

```python
df1.columns
```

```python
df1 = df1.rename(columns={'st': 'state'})
```

```python
df2.columns
```

```python
df2 = df2.rename(columns={'st': 'state'})
```

```python
df3.columns
```

```python
df1 = df1[['customer', 'state', 'customer_lifetime_value', 'gender', 'education', 'income', 'monthly_premium_auto', 'policy_type', 'number_of_open_complaints', 'vehicle_class', 'total_claim_amount']]
df2 = df2[['customer', 'state', 'customer_lifetime_value', 'gender', 'education', 'income', 'monthly_premium_auto', 'policy_type', 'number_of_open_complaints', 'vehicle_class', 'total_claim_amount']]
df3 = df3[['customer', 'state', 'customer_lifetime_value', 'gender', 'education', 'income', 'monthly_premium_auto', 'policy_type', 'number_of_open_complaints', 'vehicle_class', 'total_claim_amount']]
```

```python
df_concatenated = pd.concat([df1, df2, df3], axis=0, ignore_index=True)
df_concatenated
```

```python
numerical_columns = df_concatenated.select_dtypes(include=['number']).columns
print("Numerical Columns:", numerical_columns)
```

```python
categorical_columns = df_concatenated.select_dtypes(include=['object']).columns
print("Categorical Columns:", categorical_columns)
```

```python
print(df_concatenated.dtypes)
```

```python
df_concatenated = df_concatenated.drop(['education', 'number_of_open_complaints'], axis=1)
```

```python
df_concatenated['customer_lifetime_value'] = df_concatenated['customer_lifetime_value'].astype(str).str.replace('%', '')
```

```python
df_concatenated['customer_lifetime_value'] = df_concatenated['customer_lifetime_value'].astype(float) * 100
```

```python
df_concatenated = df_concatenated.drop_duplicates()
```

```python
df_concatenated = df_concatenated[df_concatenated['income'] > 0]
```
